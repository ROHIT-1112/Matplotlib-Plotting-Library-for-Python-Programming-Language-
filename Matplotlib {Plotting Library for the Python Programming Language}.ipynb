##### Plotting Multiple Graphs in One Visualization/Graph.

```python
experience = [1,3,5,4,7,9,11]
data_analyst_salary = [10000,35000,55000,40000,75000,90000,110000]
software_salary = [8000,38000,52000,45000,70000,85000,112000]
```

##### Pie Chart (Used for Composition) 
* A Pie Chart is a circular statistical plot that can display only one series of data. The area of the chart is the total percentage of the given data. The area of slices of the pie represents the percentage of the parts of the data.
* Official Documentation of Pie Plot/Chart :- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html
* Best Resource to Understand Pie Plot :-  https://www.geeksforgeeks.org/plot-a-pie-chart-in-python-using-matplotlib/

* Creating Pie Chart 

```python
plt.title("Pie Chart Example")
slices = [40,40,10,10]
plt.pie(slices)
plt.tight_layout()
plt.show();
```

* We can also add labels to our Pie Charts.

```python
plt.title("Pie Chart Example")
slices = [40,40,10,10]
labels_pie = ["Amy","Christy","Pollock","Brian"]
plt.pie(slices,labels =labels_pie )
plt.tight_layout()
plt.show();
```

* By default the plotting of the first wedge starts from the x-axis and move counterclockwise

* If we want to display a wedge a little bit outside of Pie Chart. Then we can use explode Parameter in Pie Chart.It takes with one value for each wedge.

* For more **Information @Explode** :- https://www.geeksforgeeks.org/radially-displace-pie-chart-wedge-in-matplotlib/

```python
plt.title("Pie Chart Example")
slices = [40,40,10,10]
exp = [0,0,0.2,0]
labels_pie = ["Amy","Christy","Pollock","Brian"]
plt.pie(slices,labels =labels_pie,explode = exp )
plt.tight_layout()
plt.show();
```

* We can also change the figsize of the figure using figsize(). 
* And can also get best dimensions using tight_layout().

```python
# Using Figsize().
plt.figure(figsize=(10,20))
plt.title("Pie Chart Example")
slices = [40,40,10,10]
exp = [0,0,0.2,0]
labels_pie = ["Amy","Christy","Pollock","Brian"]
plt.pie(slices,labels =labels_pie,explode = exp )
plt.show();
```

* We can also add Shadows to our Pie Chart.

```python
plt.title("Pie Chart Example")
slices = [40,40,10,10]
exp = [0,0,0.2,0]
labels_pie = ["Amy","Christy","Pollock","Brian"]
plt.pie(slices,labels =labels_pie,explode = exp,shadow=True )
plt.tight_layout()
plt.show();
```

* We can also add legends to our Pie Chart.


```python
plt.figure(figsize=(10,20))
plt.title("Pie Chart Example")
slices = [40,40,10,10]
exp = [0,0,0.2,0]
labels_pie = ["Amy","Christy","Pollock","Brian"]
plt.pie(slices,labels =labels_pie,explode = exp,shadow=True )
plt.legend(labels =labels_pie,loc = 'upper left')
plt.show();
```

* We can also add Percentage to our Pie Chart.

```python
plt.figure(figsize=(10,20))
plt.title("Pie Chart Example")
slices = [40,40,10,10]
exp = [0,0,0.2,0]
labels_pie = ["Amy","Christy","Pollock","Brian"]
plt.pie(slices,labels =labels_pie,explode = exp,shadow=True,autopct='%1.1f%%' )
plt.legend(labels =labels_pie,loc = 'upper left')
plt.show();
```
